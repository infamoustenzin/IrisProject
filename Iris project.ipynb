```python
#import the required packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get the dataset from sklearn
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X_iris = pd.DataFrame(iris['data'])
y_iris = pd.DataFrame(iris['target'])
```

```python
X_iris.head()
```

```python
#adding the column names
X_iris.columns = ['Sepal Length', 'Sepal Width', 'Petal Length','Petal Width']
y_iris.columns = ['class']
```

```python
print(X_iris.head())
print(y_iris.head())
```

## Exploratory Data Analysis

Since this is a classification, lets check if we have enough sample data from each class.

```python
print(y_iris['class'].value_counts())
```

```python
#check for any missing values
print(X_iris.isnull().sum())
```

```python
dataset = pd.concat([X_iris,y_iris], axis = 1)
```

```python
sns.set(style="darkgrid")
sns.relplot(x = 'Sepal Width', y = 'Sepal Length', hue = 'class', data = dataset)
plt.show()
```

```python
sns.relplot(x = 'Petal Width', y = 'Petal Length', hue = 'class', data = dataset)
plt.show()
```

From the above two visualizations, we can guess that the feature ['Petal Length', 'Petal Width'] can classify these species better than the other two
